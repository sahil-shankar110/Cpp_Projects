```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
r = Ridge(alpha=100)
```

```python
r.fit(x,y)
```

```python
print(r.coef_)
print(r.intercept_)
```

```python
r2 = Ridge(alpha=10)
r2.fit(x,y)
print(r2.coef_)
print(r2.intercept_)
```

```python
plt.plot(x,y,"b. ")
plt.plot(x,lr.predict(x),color = "black",label = "alpha 0")
plt.plot(x,r.predict(x),color = "green",label = "alpha 100")
plt.plot(x,r2.predict(x),color = "red",label = "alpha 10")
plt.legend()
plt.show()
```

# My_Ridge class

```python
class My_Ridge:
  def __init__(self,alpha=10):
    self.m = None
    self.b = None
    self.alpha = alpha

  def fit(self,x_train,y_train):
    num = 0
    den = 0

    for i in range(x_train.shape[0]):
      num = num + ((y_train[i] - y_train.mean()) * (x_train[i] - x_train.mean()))
      den = den + (x_train[i] - x_train.mean()) * (x_train[i] - x_train.mean())

    self.m = num/(den + self.alpha)
    self.b = y_train.mean() - (self.m * x_train.mean())
    print(self.m,self.b)
```

```python
ridge = My_Ridge(alpha=100)
```

```python
ridge.fit(x,y)
```
